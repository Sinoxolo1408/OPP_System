#### Importing the data

Our data was imported using the numpy function ```genfromtxt('file.csv')``` which reads and converts ```.csv``` files into a single ```numpy array``` which will be useful for dividing and saving purposes.

````python
### Importing some useful packages
#This ensures that plots appear in the sheet and not in a new window
%matplotlib inline 

#Pandas is an alternative numerical processing library like numpy: they have differences
#We will use Pandas only to provide a preview of the data once we've read it in
import pandas as pd


data = np.genfromtxt('online_shoppers_intention_selection.csv',delimiter=",")

#Convert the numpy array to a pandas DataFrame
#and display the first 10 rows as html
display(pd.DataFrame(data).head(10))
````

#### Randomizing the data 

It is important to randomize ones data set before dividing the data into Training, Cross Validtion and Testing data sets. Thus we made use of the numpy function ```numpy.random.shuffle(arr)``` to shuffle our data by rows while leaving the contents of each row unshifted.

````python
data2 =data
np.random.shuffle(data2)
data2
display(pd.DataFrame(data2).head(10))
````

#### Dividing the data

We have divided our origional data into 3 datasets which are namely, ```Training``` consisting of 60% of the origional data, ```Cross Validation``` consisting of 20% and ```Testing``` consisting of 20% of remaining origional data. 

````python
#sets the ratio of training data: validation data: testing data(currently 80:10:10)

perc1= int(0.6*len(data))
perc2= int(0.8*len(data))

train = data2[:perc1]
val = data2[perc1:perc2]
test = data2[perc2:]
````

````python
print("\nTheres a total of "+(str)(len(data2))+ " records in this data set")

#Convert the numpy array to a pandas DataFrame
#and display the first 10 rows as html for each data set

print("\nTraining data: "+(str)(len(train)))
display(pd.DataFrame(train).head(10))
print("\nCV data: "+(str)(len(val)))
display(pd.DataFrame(val).head(10))
print("\nTesting data: "+(str)(len(test)))
display(pd.DataFrame(test).head(10))
````

#### Saving the datasets

We then had to save our datasets into their respective ```.dat``` files, which was made possible through the use of the function ```savetxt('file.dat')```

````python
#writes the arrays into their respective files
savetxt('group6_train.dat', train)
savetxt('group6_cv.dat', val)
savetxt('group6_test.dat', test)
````
